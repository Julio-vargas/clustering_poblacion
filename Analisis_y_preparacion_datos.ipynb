# Preparación de los datos
### En esta sección se lleva a cabo el procesamiento de los datos para ser utilizados en los modelos de aprendizaje automático

```python
#metodos que serán de utilidad para el tratamiento de los datos:

#normaliza un dataframe utilizando MaxAbsScaler
def normalizar_dataFrame(data_normalizar):
    scaler = preprocessing.MaxAbsScaler()
    datos_normalizados = data_normalizar.copy()
    for column in datos_normalizados:
        datos_normalizados[column] = scaler.fit_transform(datos_normalizados[[column]])
    return datos_normalizados
```
